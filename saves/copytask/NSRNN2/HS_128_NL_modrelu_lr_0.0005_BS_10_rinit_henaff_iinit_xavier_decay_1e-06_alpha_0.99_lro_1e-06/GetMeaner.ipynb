```python
import numpy as np
import math
import pickle
import gzip
import pandas
import matplotlib.pylab as plt
%matplotlib inline
```

```python
loss = [] 
for i in range(1,4):
    path = "./"+str(i)+"/NSRNN2_Train_Accuracy"
    data = pickle.load(open(path, 'rb'))
    #print(data)
    loss.append(data)
avg = np.mean(loss,axis=0)
print(avg)
print("len="+str(len(avg)))
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8))
ax.plot([i for i in range(1,801)],avg,'r')
ax.set_xlabel("Epochs", fontsize=16)
ax.set_ylabel("loss", fontsize=16)
ax.legend()
```
